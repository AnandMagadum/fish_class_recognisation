```python
import cv2
import os

def load_images(folder):
    images = []
    for filename in os.listdir(folder):
        img = os.path.join(folder,filename)
        if img.endswith(".png"):
            images.append(img)
    return images
```

```python
final_image_dict={'Black_Sea_Sprat':load_images("Fish_data_setfinaal/Black_Sea_Sprat"),
                  "Gilt-Head Bream":load_images("Fish_data_setfinaal/Gilt-Head Bream"),
                  "Hourse Mackerel":load_images("Fish_data_setfinaal/Hourse Mackerel"),
                  "Red Mullet":load_images("Fish_data_setfinaal/Red Mullet"),
                  "Red Sea Bream":load_images("Fish_data_setfinaal/Red Sea Bream"),
                  "Sea Bass":load_images("Fish_data_setfinaal/Sea Bass"),
                  "Shrimp":load_images("Fish_data_setfinaal/Shrimp"),
                  "Striped Red Mullet":load_images("Fish_data_setfinaal/Striped Red Mullet"),
                  "Trout":load_images("Fish_data_setfinaal/Trout")
                
            }

image_class={"Black_Sea_Sprat":0,
            "Gilt-Head Bream":1,
            "Hourse Mackerel":2,
            "Red Mullet":3,
            "Red Sea Bream":4,
            "Sea Bass":5,
            "Shrimp":6,
            "Striped Red Mullet":7,
            "Trout":8}
```

```python
x=[]
y=[]
for images_class,images in final_image_dict.items():
    for image in images:
        image=cv2.imread(image)
        resized_image=cv2.resize(image,(400,400))

        x.append(resized_image)
        y.append(image_class[images_class])
    
    
```

```python
len(x),len(y)
```

```python
x=np.array(x)
```

```python
y=np.array(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,stratify=y,test_size=0.25)
```

```python
model=keras.Sequential([
     keras.layers.Conv2D(16,(3,3),padding='same',input_shape=(400,400,3),activation='relu'),
     keras.layers.MaxPool2D(2,2),
     keras.layers.Flatten(),
     keras.layers.Dense(64,activation='relu'),
     keras.layers.Dense(108,activation='relu'),
     keras.layers.Dense(9,activation='sigmoid')                    
])
```

```python
model.compile(optimizer='Adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'],)
```

```python
model.fit(x_train,y_train,epochs=10)
```

```python
model.evaluate(x_test,y_test)
```
